# 一个例子：关系最好的两个人是谁？

```python
df = pd.merge(enrollment,students,on='student_id')
relation = {}
for item in df.groupby(['course_id','term']).apply(lambda x:sorted(x['name'].values.tolist())):
    for i,a in enumerate(item):
        for b in item[i+1:]:
            relation[a+'&'+b]=relation.get(a+'&'+b,0)+1
```

```python
res = pd.DataFrame(list(relation.items()),columns=['relation','count'])
```

```python
res[res['count']==res['count'].max()]
```
